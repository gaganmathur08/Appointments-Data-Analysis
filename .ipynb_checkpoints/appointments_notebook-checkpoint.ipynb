### Data Cleaning 

        1) After looking at the dataset it looks like the field 'ScheduledDay' and 'AppointmentDay' datetime value contains T,Z characters in between date and time. So we need to clean it and return the date in datetime format.
        2) No-show field contains Yes and No which are not easily comparable in if statement so we will convert it into boolean value.
        3) Record constains duplicate patient ids. So we need to get unique records.

```python
#1) Clean datetime. convert it into datetime format.

def parse_date(date):
    if date == '':
        return None
    else:
        return dt.strptime(date,'%Y-%m-%dT%H:%M:%SZ')
    
```

```python
#2) Clean No-Show field. Make no_show return boolean values. i.e. True for No and False for Yes.

def clean_no_show(value):
    return value == 'No'

appointments_df['No-show'] = appointments_df['No-show'].apply(clean_no_show)
appointments_df['ScheduledDay'] = appointments_df['ScheduledDay'].apply(parse_date)
appointments_df['AppointmentDay'] = appointments_df['AppointmentDay'].apply(parse_date)

#print(appointments_df.head())
```

```python
#Lets remove duplicates from the dataset.

unique_appointment_df = appointments_df.drop_duplicates(subset=['PatientId'],inplace=False)

```

## Exploratory Data Analysis

### 1) Which age group have more patients of diabetes

     Firstly,By using pearson's r we can find the correlation between the age,Neighbourhood and diabetes variables.
     If pearson's r value is large that means there is a strong relationship between the variables.


```python
#For calculating pearsons r value between age and diabetes.
def find_correlation_r(x,y):
    std_x = (x-x.mean())/x.std(ddof=0)
    std_y = (y-y.mean())/y.std(ddof=0)
    
    return (std_x * std_y).mean()

print(find_correlation_r(unique_appointment_df['Age'],unique_appointment_df['Diabetes']))
```

    Here correlation value between age and diabetes is 0.2923914636778956. Which means both are positive. So we can say that 
    if the age of patients increases then the diabetes patient also increases.

```python
# Set figure width to 12 and height to 9
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 6
fig_size[1] = 4
plt.rcParams["figure.figsize"] = fig_size

#Plot distribution for age vs diabetes
plt.hist(unique_appointment_df['Age'])
plt.title('Number of patients by age')
plt.xlabel('Ages')
plt.ylabel('Number of patients')
plt.show()
```

    Here from the above distribution we can see that most of the patients are of age between 0 to 70. And maximum patients are of age group between 0 to 10.
    
### Lets create age groups as per following criteria:
    Children upto 9 years of age
    Adolescents 10-19 years
    Adults 20-45 years
    Middle age 46-60 years
    Old > 60 

```python
#function for categorizing data into age groups

def convert_to_age_group(age):
    if age <= 9:
        return 'Childeren'
    elif age <=19:
        return 'Adolscents'
    elif age <= 45:
        return 'Adults'
    elif age <= 60:
        return 'Middle age'
    else:
        return 'Old'
    
unique_appointment_df.is_copy = False
unique_appointment_df['Age-Group'] = unique_appointment_df['Age'].apply(convert_to_age_group)
unique_appointment_df.head()
```

```python
#To find which age group have more diabetes patients first group data based on age groups.
#Then sum total number of diabetes patient in each group.
appointments_by_agegroup = unique_appointment_df.groupby('Age-Group')
appointments_by_agegroup['Diabetes'].sum()
```

    By looking at the above values it is clear that Old age patients have more number of diabetes. Based on this we can conclude that age is a factor based on which we can say that patient can have diabetes.
    
    Let's also visualize this in graph.

```python
ax = appointments_by_agegroup['Diabetes'].sum().plot(title='Number of patients based on age groups')
ax.set_xlabel('Age groups')
ax.set_ylabel('Number of patients')
```

### 2) What days patients mostly prefer for appointment (Weekdays or weekends).

    To identify this point we first need to add one more column in our dataset for day of week. This will help us to perform analysis based on day of week.

```python
def get_week(datestr):
    return dt.weekday(datestr)+1

unique_appointment_df['day_of_week'] = unique_appointment_df['ScheduledDay'].apply(get_week)
unique_appointment_df.head()
#Sunday = 7,Monday = 1
```

```python
#For calculating days on which patients mostly prefer appointments,
#First group data by day_of_week

appointments_by_dow = unique_appointment_df.groupby('day_of_week')
print(appointments_by_dow.size())
ax = appointments_by_dow.size().plot(title='Patients appointment based on day')
ax.set_ylabel('Number of patients')
```

    By looking at the above graph we can say that on Sundays no one has scheduled the appointment. On saturday very few patients have scheduled appointments. Maximum appointments are on week 2 i.e Tuesday.

```python
#But not all the patients have showed up on their scheduled day.
#So to get the actual number of patients who showed up for their appointment,
#Lets just calculate count those rows where No-Show is False.

print(appointments_by_dow['No-show'].sum())
ax = appointments_by_dow['No-show'].sum().plot(title = 'Number of patients who really showed up on the scheduled appointment day')
ax.set_ylabel('Number of patients')
```

    Here distribution looks similar to the distribution for all the scheduled appointments. So based on the above distributions we can say that maximum appointments are scheduled on Tuesday. Very few appointments on Saturday, and on Sunday's there is no scheduled appointment. 
    Patients mostly prefer weekdays for appointments.

### 3) Which neighbourhood is more alchol addicted?

    For analyzing this, first lets see whether there is any relation between the two columns or not.

```python
#We have already defined a function for calculating pearson's r between two variables.
```

```python
#For caluculating alchol effect for each neighbourhood we first need to group our data based on neighbourhood.

appointments_by_nb = unique_appointment_df.groupby('Neighbourhood',as_index=False)
```

```python
#After grouping lets find total number of alchol patients in each neighbourhood!
#Then plot the bar chart to visualize the results.
sum_appointment_by_nb_alchol = appointments_by_nb['Alcoholism'].sum()

neighbourhoods = list(sum_appointment_by_nb_alchol['Neighbourhood'])
sum_alcholism = list(sum_appointment_by_nb_alchol['Alcoholism'])



#Reference for setting seaborn https://datahub.packtpub.com/tutorials/visualizing-univariate-distribution-seaborn/
snb.set(rc={"figure.figsize": (15, 8)})
snb.barplot(x='Neighbourhood',y='Alcoholism',data=sum_appointment_by_nb_alchol)
# Use Matplotlib functions to label axes rotate tick labels
ax = plt.gca()
ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
plt.show()


```

```python
#To directly get the neighbourhood where alcholic patients are more lets create a function

def get_max_alcholic_neighbour(data):
    max_index = sum_appointment_by_nb_alchol['Alcoholism'].argmax()
    return sum_appointment_by_nb_alchol['Neighbourhood'].iloc[max_index]


print('Neighbourhood consumes maximum amount of alcohol: {}'.format(get_max_alcholic_neighbour(sum_appointment_by_nb_alchol)))
```

    By looking at the above results it is clear that 'SANTA MARTHA' consumes maximum amount of alchol.
    
### 4) Are diabetes patients consuming alchol?

## 7) Which patient is going for appointment more often?

```python

appointments_by_patients = appointments_df.groupby('PatientId',as_index=False)
sum_appointment_by_patients = appointments_by_patients.size().reset_index(name='counts')

def get_max_appointment_patient_id(data):
    max_index = data['counts'].argmax()
    return data['PatientId'].iloc[max_index]



sum_appointment_by_patients.loc[sum_appointment_by_patients['PatientId'] == get_max_appointment_patient_id(sum_appointment_by_patients)]
```
